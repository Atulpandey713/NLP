```python
corpus="""Hello Welcome ,to Atul Pandey's NLP Tutorials.
Please do learn and read very carefull ! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
## Toknization
## Sentence -->paragraphs
from nltk.tokenize import sent_tokenize
```

```python
Doc=sent_tokenize(corpus)
```

```python
Doc
```

```python
type(Doc)
```

```python
for sentence in Doc:
    print(sentence)
```

```python
## Tocknization
## Paragraphs --> words
## sentence --> words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in Doc:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer=TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```
